```python
import tkinter as tk

# 画面作成
tki = tk.Tk()
tki.geometry('300x200')
tki.title('ﾊｰﾄﾞﾗｲﾄの合成色を求める')

# ラベル
X,Y=35,30 # ラベルやテキストボックスの位置を調整しやすいように設定した変数……(0)

lbl_1 = tk.Label(text='基本色 : R')
lbl_1.place(x=X,y=Y)
lbl_2 = tk.Label(text='G')
lbl_2.place(x=X+85,y=Y)
lbl_3 = tk.Label(text='B')
lbl_3.place(x=X+125,y=Y)

Y += 30
lbl_4 = tk.Label(text='結果色 : R')
lbl_4.place(x=X,y=Y)
lbl_5 = tk.Label(text='G')
lbl_5.place(x=X+85,y=Y)
lbl_6 = tk.Label(text='B')
lbl_6.place(x=X+125,y=Y)

Y += 50
lbl_7 = tk.Label(text='合成色 : R')
lbl_7.place(x=X,y=Y)
lbl_8 = tk.Label(text='G')
lbl_8.place(x=X+85,y=Y)
lbl_9 = tk.Label(text='B')
lbl_9.place(x=X+125,y=Y)

#テキストボックス
X,Y=35,30 #(1)

txt_1 = tk.Entry(width=3)
txt_1.place(x=X+60,y=Y)
txt_2 = tk.Entry(width =3)
txt_2.place(x=X+100,y=Y)
txt_3 = tk.Entry(width =3)
txt_3.place(x=X+140,y=Y)

Y += 30 #(2)
txt_4 = tk.Entry(width=3)
txt_4.place(x=X+60,y=Y)
txt_5 = tk.Entry(width =3)
txt_5.place(x=X+100,y=Y)
txt_6 = tk.Entry(width =3)
txt_6.place(x=X+140,y=Y)

Y += 50 #(3)
txt_7 = tk.Entry(width=3)
txt_7.place(x=X+60,y=Y)
txt_8 = tk.Entry(width =3)
txt_8.place(x=X+100,y=Y)
txt_9 = tk.Entry(width =3)
txt_9.place(x=X+140,y=Y)

# 下の(4)の計算で用いる変数
base = [0.0]*3 # 基本色(float)
result = [0.0]*3 # 結果色(float)
ibase = [0]*3 # 基本色(int)
iresult = [0]*3
icom = [0]*3 # 合成色(int)
v0,v1,v2 = 0,0,0
S,M = 255.0,0.0 # S, Mはそれぞれスクリーン風、乗算風に計算された場合の値v0。ただしfloat

# クリックイベントの定義
def btn_click():
    # テキストボックスのクリア
    txt_7.delete(0, tk.END)
    txt_8.delete(0, tk.END)
    txt_9.delete(0, tk.END)

    # テキスト取得
    ibase[0] = int(txt_1.get())
    ibase[1] = int(txt_2.get())
    ibase[2] = int(txt_3.get())
    iresult[0] = int(txt_4.get())
    iresult[1] = int(txt_5.get())
    iresult[2] = int(txt_6.get())

    # 合成色を計算する……(4)
    for i in range(3):
        base[i] = float(ibase[i])
        result[i] = float(iresult[i])

        # S, Mを計算
        if base[i] == 255:
            if result[i] == 255:
                S = 255 # 理屈としてはなんでもいいが、多分これが塗ってて一番違和感のない数値
            else:
                S = -56562 # ⇒S < float(255)/2 ⇒ 乗算風に計算
        else:
            S = (255*(-2*base[i] + result[i] + 255))/(2*(255 - base[i]))
            M = 255*result[i]/(2*base[i])

        # 合成色のリストicomの各項に整数値を代入
        if S >= float(256)/2:
             v0 = S
        else:
            v0 = M
        v1 = float(int(v0))
        v2 = (v0 -v1)*10-5
        v2 = int(v0) + int(v2 >= 0) # v0の小数点第一位を四捨五入
        icom[i] = v2

    # 値をテキストボックスへセット
    txt_7.insert(0,icom[0])
    txt_8.insert(0,icom[1])
    txt_9.insert(0,icom[2])

    # 結果をprint
    o = "基本色=" + str(ibase) + ", 合成色=" + str(icom) +"⇒結果色=" + str(iresult)
    print(o)

# ボタン作成
btn = tk.Button(tki, text='計算', command = btn_click)
X += 170 #cf.(1),(2),(3)
Y -= 55
btn.place(x=X, y=Y)
```
